```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
df_train.head()
```

```python
df_group_train = df_train['PassengerId'].str.split('_', n=1, expand=True).rename(columns={0: 'Group', 1:'Id'})
df_train['Group'], df_train['IdPassengerInGroup'] = df_group_train['Group'], df_group_train['Id']
df_cabin_train = df_train['Cabin'].str.split('/', expand=True).rename(columns={0: 'Deck', 1:'Number', 2:'Side'})
df_train['Deck'], df_train['Number'], df_train['Side'] = df_cabin_train['Deck'], df_cabin_train['Number'], df_cabin_train['Side']

df_group_test = df_test['PassengerId'].str.split('_', n=1, expand=True).rename(columns={0: 'Group', 1:'Id'})
df_test['Group'], df_test['IdPassengerInGroup'] = df_group_test['Group'], df_group_test['Id']
df_cabin_test = df_test['Cabin'].str.split('/', expand=True).rename(columns={0: 'Deck', 1:'Number', 2:'Side'})
df_test['Deck'], df_test['Number'], df_test['Side'] = df_cabin_test['Deck'], df_cabin_test['Number'], df_cabin_test['Side']
```
